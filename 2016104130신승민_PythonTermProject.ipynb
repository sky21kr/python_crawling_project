영화 '인터스텔라' 개봉 당시 영화 '패션왕'이 개봉했었다. 인터스텔라는 당시 천만관객을 돌파하였고 패션왕은 손익분기점이 130만 관객 이였는데 59만 관객에서 그치고 말았다. 그래서 그 시기에 나왔던 말이 '명예로운 죽임을 당했다'였다. 이 뜻은 패션왕이 인터스텔라와 개봉 시기가 겹치지 않았어도 망했을 영화라는 것이다. 만화가 이말년도 이것을 보고 위와 같은 그림을 그렸었다.

 그럼 과연 '인터스텔라' 같은 천만 관객 영화가 주변 영화 즉 같이 상영 중인 영화들에 영향을 어떤 미쳤을까? 이 점이 궁금하여 이 주제를 선정하게 되었다. 그래서 천만 관객 영화가 개봉한 시기에 같이 상영 중인 영화들의 누적 관객수를 기준으로 하여 주변 영화에 미친 영향을 조사할 것이다.
 
## (2) 가설 정의

- 천만 관객 영화가 주변 영화들의 흥행을 방해할 것이다.

 가설을 증명하기 위해 2019년 11월 3일부터 2009년 1월1일까지 박스 오피스 순위에 올라와 있었던 영화들의 정보를 받아올 것이다. 받아올 영화의 정보는 영화의 제목, 영화의 개봉년도, 영화의 누적 관객 수 이다. 그리고 천만 관객 영화의 개봉일을 기준으로 2주 전, 1주 전, 같은 주, 1주 뒤, 2주 뒤를 1주일 씩 기간으로 잡아서 그 기간동안 개봉한 영화들의 누적 관객수의 평균 값들을 구하여 전체 영화의 누적 관객수와 비교할 것이다.

## (6) 결론: 본인의 명제/논리가 증명/확인 되었는지와 이를 구체적으로 서술함
22편의 영화 중 국제시장을 제외한 21편의 영화에서 같은 주에 개봉한 영화의 누적 관객수가 평균보다 적었다. 약 95%의 확률로 같은 주에 개봉한 영화는 흥행에 실패했다고 볼 수 있다. 그런데 개봉 이후에 서서히 증가하는 경우는 22편 중 6편 밖에 나오지 않았다. 그럼 다른 주는 어떨까?

2주 전 주에서 평균보다 낮은 영화는 22편의 영화 중 16편, 1주 전 주는 11편, 1주 후는 9편, 2주 후는 13편 이였다. 같은 주 다음으로 낮을 것으로 예상하였던 1주 후 주의 영화가 제일 흥행률이 좋았다. 이것을 통해 천만 관객 영화가 같은 주에 개봉한 영화의 흥행에 부정적인 영향을 미친 것은 확실하지만 그 외 나머지 주에도 부정적인 영향을 미쳤다고 보기는 어렵다. 오히려 1주 뒤 주의 영화들은 약 60%의 확률로 평균을 넘었고 평균을 넘은 13편 중 11편이 평균과 2배 이상의 큰 차이를 보여주는 것으로 보아 흥행에 긍정적인 영향을 미쳤다고 볼 수도 있다. 이에 정확한 결과를 도출해내기 위해서는 다른 조건의 실험이 필요하다고 볼 수 있다.

정리하자면 천만 관객 영화는 같은 주에 개봉한 영화의 흥행을 방해한다. 하지만 그 외의 주에 대해서는 방해했다고 말하기 어렵고 오히려 이후에 개봉한 영화에는 흥행에 긍정적인 영향을 미쳤다고 볼 수 있다. 

결론: 패션왕은 인터스텔라가 망하는데 한 몫 했다고 볼 수 있고 같은 날이 아닌 그 다음주에 개봉 했었더라면 더 좋은 성적을 거둘 수 있었을 것이다.

번외: 어벤져스 시리즈와 인터스텔라가 개봉한 시기 근처에 개봉한 영화들은 5주 모두 평균을 넘지 못했다. 이 영화들의 공통점은 대중적이고 개봉 전부터 유명했다는 것이다. 더 자세한 결과를 얻기 위해서는 마블 영화로 실험하는 등의 더 많은 표본이 필요하겠지만 망하고싶지 않다면 흥행이 보장된 영화의 개봉날 근처는 피하는게 좋다고 생각한다.
